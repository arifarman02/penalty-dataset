```python
# Import pandas and load data as a dataframe
import pandas as pd

df = pd.read_csv("penalty_dataset.csv")

df
```

```python
# Convert match data to datetime
df['match_date'] = pd.to_datetime(df["match_date"])
```

```python
# Who has scored the 5th most penalties since the start of 2015 (excluding penalty shootouts)? And how many have they scored
penalties_since_15 = df[(df["match_date"].dt.year >= 2015) & (df["event_type"] == "Penalty Goal") & (df["shootout_order"].isna())]

player_penalties = penalties_since_15["player"].value_counts().reset_index()
player_penalties.columns = ["player", "penalty_count"]

if len(player_penalties) >= 5:
    player = player_penalties.iloc[4]
    print(player)
```

```python
# What 10 minute window had the most penalties awarded within it and how many penalties were awarded?
df_penalties = df[df["event_type"].str.contains("Penalty") & ~df["event_type"].str.contains("Shootout")].dropna(subset=['time'])

# Create time windows
df_penalties.loc[:,"time_window"] = (df_penalties["time"] // 10) * 10
time_window_counts = df_penalties["time_window"].value_counts().sort_index()
if not time_window_counts.empty:
    most_common_window = time_window_counts.idxmax()
    most_common_count = time_window_counts.max()
    print(f"{int(most_common_window)}-{int(most_common_window)+10}")
    print(most_common_count)
else:
    print("No penalties found in the dataset with valid time values.")
```
